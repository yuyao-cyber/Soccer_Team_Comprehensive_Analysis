### MW1 Team 0

```python
import warnings
warnings.filterwarnings("ignore")

#calculate each framecount in minutes
team_0['Minutes_Played'] = 1/30/60

#Assume the instant speed is constant. Calculate the distance in each frame.
team_0['Total_Distance'] = team_0['PlayerSpeed']/30

#As a reference, according to the article 
#https://www.frontiersin.org/articles/10.3389/fspor.2023.1116293/full#:~:text=During%20official%20matches%2C%20high%2Dspeed,in%20professional%20male%20soccer%20players.
#high-speed distance ranges from 19.8km/h ~ 25.1km/h which is 5.5m/s ~ 7.0m/s
#and sprint distance ranges over 25.1km/h which is greater than 7.0m/s.

#Use distance = speed * time. Find PlayerSpeed within the range of 5.5 and 7, then divide it by 30 to get distance
#per FrameCount
team_0['High_Speed_Distance'] = team_0.loc[(team_0['PlayerSpeed']>5.5) & (team_0['PlayerSpeed']<7),'PlayerSpeed']/30
#Fill the None value in 'High_Speed_Distance' with 0.
team_0['High_Speed_Distance'].fillna(0,inplace=True)

#Use distance = speed * time. Find PlayerSpeed greater than 7, then divide it by 30 to get distance
#per FrameCount
team_0['Sprint_Distance'] = team_0.loc[team_0['PlayerSpeed']>=7,'PlayerSpeed']/30
#Fill the None value in 'Sprint_Distance' with 0.
team_0['Sprint_Distance'].fillna(0,inplace=True)
team_0
```

```python
#The logic to find Sprint_Count is to divide FrameCount into groups with consecutive numbers. For each group,
#if there appears PlayerSpeed grater than 7, Sprint_Count += 1. 

# initializing list
tuple_list = []
#Set groups to divide FrameCount. If FrameCount.diff is greater than 1, set to another group.
team_0['Sprint_Count_Group'] = (team_0['FrameCount'].diff()>1).cumsum()
sprint_count_group = team_0.groupby(['Sprint_Count_Group','JerseyNumber'])
for name, group in sprint_count_group:
    #Search if any PlayerSpeed is greater than 7
    if (group['PlayerSpeed'] >=7).any():
        tuple_list.append(name)

result = defaultdict(list)
for sub in tuple_list:
    # getting all keys to values

    result[sub[1]].append(sub[0])
#put result into dict object
result = dict(result)
res_dict = dict()
for key in result:
    # getting unique key counts for each value
    res_dict[key] = len(list(set(result[key])))

res_df = pd.DataFrame(list(res_dict.items()),columns=['JerseyNumber','Sprint_Count'])

#Select columns from team_0
team0_result = team_0.groupby('JerseyNumber')[['Minutes_Played','Total_Distance','High_Speed_Distance','Sprint_Distance']].sum()
#Reset Indices to merge
team0_new = team0_result.reset_index()
#Merge two df
team0_final = team0_new.merge(res_df,how='left')
#fill Sprint_count = None with 0
team0_final = team0_final.fillna(0)
team0_final
```

## Pressing Strategy Analysis

```python
def find_game_frame_count(df, goal_keeper_jersey_0,goal_keeper_jersey_1):
    df_team0 = df.loc[(df['TeamType']== 0)]
    df_team0_goalkeeper = df_team0.loc[df_team0['JerseyNumber']==goal_keeper_jersey_0]
    temp1 = df_team0_goalkeeper.loc[df_team0_goalkeeper['XPosition']>0].sort_values(by='FrameCount')
    temp2 = df_team0_goalkeeper.loc[df_team0_goalkeeper['XPosition']<0].sort_values(by='FrameCount')
    if temp1['FrameCount'].max() < temp2['FrameCount'].min():
        first_half_0 = df_team0.loc[(df_team0['FrameCount']>=temp1['FrameCount'].min()) & (df_team0['FrameCount']<=temp1['FrameCount'].max())]
        second_half_0 = df_team0.loc[(df_team0['FrameCount']>=temp2['FrameCount'].min()) & (df_team0['FrameCount']<=temp2['FrameCount'].max())]
        first_half_0_attack = first_half_0.loc[first_half_0['XPosition']<0]
        second_half_0_attack = second_half_0.loc[second_half_0['XPosition']>0]

    else:
        second_half_0 = df_team0.loc[(df_team0['FrameCount']>=temp1['FrameCount'].min()) & (df_team0['FrameCount']<=temp1['FrameCount'].max())]
        first_half_0 = df_team0.loc[(df_team0['FrameCount']>=temp2['FrameCount'].min()) & (df_team0['FrameCount']<=temp2['FrameCount'].max())]
        first_half_0_attack = first_half_0.loc[first_half_0['XPosition']>0]
        second_half_0_attack = second_half_0.loc[second_half_0['XPosition']<0]
        
    df_team1 = df.loc[(df['TeamType']== 1)]
    df_team1_goalkeeper = df_team1.loc[df_team1['JerseyNumber']==goal_keeper_jersey_1]
    temp1 = df_team1_goalkeeper.loc[df_team1_goalkeeper['XPosition']>0].sort_values(by='FrameCount')
    temp2 = df_team1_goalkeeper.loc[df_team1_goalkeeper['XPosition']<0].sort_values(by='FrameCount')
    if temp1['FrameCount'].max() < temp2['FrameCount'].min():
        first_half_1 = df_team1.loc[(df_team1['FrameCount']>=temp1['FrameCount'].min()) & (df_team1['FrameCount']<=temp1['FrameCount'].max())]
        second_half_1 = df_team1.loc[(df_team1['FrameCount']>=temp2['FrameCount'].min()) & (df_team1['FrameCount']<=temp2['FrameCount'].max())]
        first_half_1_attack = first_half_1.loc[first_half_1['XPosition']<0]
        second_half_1_attack = second_half_1.loc[second_half_1['XPosition']>0]

    else:
        second_half_1 = df_team1.loc[(df_team1['FrameCount']>=temp1['FrameCount'].min()) & (df_team1['FrameCount']<=temp1['FrameCount'].max())]
        first_half_1 = df_team1.loc[(df_team1['FrameCount']>=temp2['FrameCount'].min()) & (df_team1['FrameCount']<=temp2['FrameCount'].max())]
        first_half_1_attack = first_half_1.loc[first_half_1['XPosition']>0]
        second_half_1_attack = second_half_1.loc[second_half_1['XPosition']<0]
    return first_half_0_attack, second_half_0_attack, first_half_1_attack, second_half_1_attack
```

```python
def is_within_range(center_x, center_y, x, y, radius=600):
    """
    Check if a given point (x, y) is within a given radius from a center point (center_x, center_y)
    """
    distance = np.sqrt((x - center_x) ** 2 + (y - center_y) ** 2)
    return distance <= radius
```
